```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Iris.csv')
```

```python
dataset.head()
```

```python
columns = [1,3]
X = dataset.iloc[:,columns].values
y = dataset.iloc[:,5].values
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder() 
```

```python
y = label_encoder.fit_transform(y)
```

```python
set(y)
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
```

```python
c=[0.01,0.5,1,5,10]
index = [1,2,3,4,5]
plt.figure(figsize=(15, 15))
for i in range(len(c)):
    print("C =",c[i])
    svm = SVC(kernel='linear', C=c[i])
    svm.fit(x_train,y_train)
    y_pred = svm.predict(x_test)
    print("accuracy : ", accuracy_score(y_test, y_pred))
    xx = np.arange(x_train[:,0].min() - 0.5, x_train[:,0].max() + 0.5, 0.01)
    yy = np.arange(x_train[:,1].min() - 0.5, x_train[:,1].max() + 0.5, 0.01)
    x1,x2 = np.meshgrid(xx,yy)
    z = svm.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape)
    for j in range(len(index)):
        plt.subplot(3,2,index[j])
        plt.contourf(x1,x2,z,cmap=ListedColormap(('cyan','y','r')))
        plt.scatter(x_train[:,0], x_train[:,1],c=y_train)
```
